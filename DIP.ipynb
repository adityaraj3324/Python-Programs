```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load image
image_path = 'DD2.jpg'  # üîÅ Change this to your image file
img = cv.imread(image_path)

# Convert to grayscale
gray = cv.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define Laplacian kernel for sharpening
laplacian_kernel = np.array([[0, -1, 0],
                             [-1, 5, -1],
                             [0, -1, 0]])

# Apply the kernel to sharpen the image
sharpened = cv.filter2D(gray, -1, laplacian_kernel)

# (Optional) Create a dummy DataFrame to show grayscale values (for educational purpose)
df = pd.DataFrame(sharpened)
print("Sample of pixel intensities after sharpening:")
print(df.head())

# Display results using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load image
image_path = 'BT2.jpg'  # Replace with your image
img = cv.imread(image_path)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Function to add Gaussian noise
def add_gaussian_noise(image, mean=0, std=25):
    noise = np.random.normal(mean, std, image.shape).astype(np.float32)
    noisy = image.astype(np.float32) + noise
    noisy = np.clip(noisy, 0, 255).astype(np.uint8)
    return noisy

# Add noise
noisy_img = add_gaussian_noise(gray)

# Apply Box Filter
box_filtered = cv.blur(noisy_img, (3, 3))

# Apply Gaussian Blur
gaussian_filtered = cv.GaussianBlur(noisy_img, (3, 3), 0)

# Apply Median Filter
median_filtered = cv.medianBlur(noisy_img, 3)

# Display sample values using pandas
df = pd.DataFrame(noisy_img)
print("Sample pixel values after adding noise:")
print(df.head())

# Show all images
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(noisy_img, cmap='gray')
plt.title('Noisy (Gaussian)')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(box_filtered, cmap='gray')
plt.title('Box Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filter')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load image and convert to grayscale
image_path = 'BT.jpg'  # Replace with your image path
img = cv.imread(image_path)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# 1. Sobel Edge Detection (X and Y)
sobel_x = cv.Sobel(gray, cv.CV_64F, 1, 0, ksize=3)
sobel_y = cv.Sobel(gray, cv.CV_64F, 0, 1, ksize=3)
sobel_combined = cv.magnitude(sobel_x, sobel_y)
sobel_combined = np.uint8(np.clip(sobel_combined, 0, 255))

# 2. Laplacian Edge Detection
laplacian = cv.Laplacian(gray, cv.CV_64F)
laplacian = np.uint8(np.clip(np.absolute(laplacian), 0, 255))

# 3. Canny Edge Detection
canny = cv.Canny(gray, 100, 200)

# Optional: Display pixel values
df = pd.DataFrame(canny)
print("Sample edge pixel values (Canny):")
print(df.head())

# Plot all edge detections
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Combined')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(canny, cmap='gray')
plt.title('Canny')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load and convert to grayscale
image_path = 'BT3.jpg'  # Replace with your image
img = cv.imread(image_path)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# 1. Simple Binary Thresholding
_, thresh_binary = cv.threshold(gray, 127, 255, cv.THRESH_BINARY)

# 2. Otsu's Thresholding (automatically determines best threshold)
_, thresh_otsu = cv.threshold(gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

# 3. Adaptive Thresholding (good for uneven lighting)
adaptive_thresh = cv.adaptiveThreshold(gray, 255,
                                       cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                                       cv.THRESH_BINARY, 11, 2)

# Display some values with pandas
df = pd.DataFrame(thresh_binary)
print("Sample thresholded pixel values (Binary):")
print(df.head())

# Plotting results
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(thresh_binary, cmap='gray')
plt.title('Simple Threshold (127)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(thresh_otsu, cmap='gray')
plt.title("Otsu's Threshold")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Adaptive Thresholding')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load image
image_path = 'BT1.jpg'  # Replace with your image path
img = cv.imread(image_path)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Apply binary threshold
_, thresh = cv.threshold(gray, 127, 255, cv.THRESH_BINARY_INV)

# Find contours (external for object outlines)
contours, _ = cv.findContours(thresh, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

# Draw and label each object
labeled_img = img.copy()
for i, contour in enumerate(contours):
    # Get bounding box
    x, y, w, h = cv.boundingRect(contour)
    cv.rectangle(labeled_img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv.putText(labeled_img, f"Obj {i+1}", (x, y - 10),
               cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

# Display results
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv.cvtColor(gray, cv.COLOR_BGR2RGB))
plt.title('Grayscale')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(thresh, cmap='gray')
plt.title('Thresholded (Segmentation)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv.cvtColor(labeled_img, cv.COLOR_BGR2RGB))
plt.title('Labeled Objects')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# Create a black image
height, width = 500, 500
black_image = np.zeros((height, width), dtype=np.uint8)

# Define font properties
font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 5
thickness = 10
font_color = 255  # White text on black background

# Write text in center
text = "Aman"
text_size = cv.getTextSize(text, font, font_scale, thickness)[0]
text_x = (black_image.shape[1] - text_size[0]) // 2
text_y = (black_image.shape[0] + text_size[1]) // 2
cv.putText(black_image, text, (text_x, text_y), font, font_scale, font_color, thickness)

# Define kernel for morphological operations
kernel = np.ones((15, 15), np.uint8)  # Large kernel for visible effects

# Apply morphological operations
erosion = cv.erode(black_image, kernel, iterations=1)
dilation = cv.dilate(black_image, kernel, iterations=1)
opening = cv.morphologyEx(black_image, cv.MORPH_OPEN, kernel)
closing = cv.morphologyEx(black_image, cv.MORPH_CLOSE, kernel)

# Titles and images for plotting
titles = ["Original", "Erosion", "Dilation", "Opening", "Closing"]
images = [black_image, erosion, dilation, opening, closing]

# Plot results
plt.figure(figsize=(10, 7))
for i in range(len(titles)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
